```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) 


# Read in the data

df = pd.read_csv(r'C:\Users\Ogunjimi\Desktop\Datasets\movies.csv')
```

```python
# To look at the data

df.head()
```

```python
# Using a for loop to check for missing data in columns

for col in df.columns:
    pct_missing = round(np.mean(df[col].isnull()), 2)
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Replace missing values with 0

df['budget'] = df['budget'].fillna(0)

df['gross'] = df['gross'].fillna(0)

df = df.fillna(0)
```

```python
# Data types of colums

df.dtypes
```

```python
# To change the data type of budget and gross to integer

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df['released'] = df['released'].astype(str)

df.dtypes
```

```python
# Create correct Year column that aligns with release date

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(str)

df
```

```python
# Sort and order data by revenue(gross)

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Correlation Analysis

# Hypothesis 1: There is a high correlation between Movie Budget and Gross earnings
# Hypothesis 2: There is a high correlation between Production Company and Gross earnings
              
```

```python
# Testing Hypothesis 1 
```

```python
# Sactter Plot with Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget Vs Gross Earnings')

plt.xlabel('Movie Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Seaborn Regression Plot for Budget Vs Gross 

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color': 'blue'})

plt.title('Budget Vs Gross Earnings')

plt.xlabel('Movie Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Actual correlation (numerical)

df.corr(method='pearson')
```

```python
# Visual representation of correlation matrix

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Hyotheses 1 is Accepted 
# There is a high correlation between budget and gross earning (pearson corrrelation co-efficient = 0.75)
```

```python
# Testing Hypothesis 2
```

```python
# Numerize non-numerical data for correlation analysis

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes

df_numerized
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) >= 0.5]

high_corr
```

```python
# Visualize high correlations

corr_mat = df_numerized.corr()

high_corr = corr_mat[((corr_mat >= .5) | (corr_mat <= -.5)) & (corr_mat !=1.000)]

sns.heatmap(high_corr, annot=True, cmap="Reds")

plt.show()
```

```python
# Hypothesis 2 is Rejected. 
# There is no significant correlation between Company and Gross earnings
```
