```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) 
```

```python
# Read in the data

df = pd.read_csv(r'C:\Users\Ogunjimi\Desktop\Datasets\Olympic_games_transformed.csv')
```

```python
df.head()
```

```python
# Assign column names

df.columns = ['ID', 'Competitor', 'Gender', 'Age', 'Age Group', 'Height', 'Weight', 'Nation Code', 'Year', 'Sport', 'Event', 'Medal']
```

```python
df
```

```python
# Drop rows with missing data in Age and Age group

data = df.dropna(subset=['Age', 'Age Group'])
```

```python
data
```

```python
# Change datatype of Age from float to integer

data['Age'] = data['Age'].astype('int64')

data.head()
```

```python
# Isolate Non registered medals from Medal column

delete = data['Medal'] == 'Not Registered'
```

```python
# Exclude Non registed medals from final dataframe for analysis

final_data = data[~delete]

final_data
```

```python
final_data.dtypes
```

```python
# Numerize the data for correlation analysis

data_numerized = final_data

for col in data_numerized.columns:
    if (data_numerized[col].dtype=='object'):
        data_numerized[col] = data_numerized[col].astype('category')
        data_numerized[col] = data_numerized[col].cat.codes


data_numerized
```

```python
correlation_matrix = data_numerized.corr()
```

```python
# Plot a heatmap showing correlations

sns.heatmap(correlation_matrix, annot=True)

plt.show()
```

```python
final_data
```

```python
# Derive dataframe for Age and Medals won

medal_age = final_data.groupby(['Age'])['Medal'].count().reset_index()

medal_age
```

```python
# Correlation Matrix for age and medals

corr_mat = medal_age.corr()

corr_mat
```

```python
# Seaborn plot for correlation

sns.heatmap(corr_mat, annot=True)

plt.show()
```

## Conclusion

#### There is a negative correlation between competitors' age and performance (using number of medals won as the performance metric) 

***Pearson correlation co-efficient = -0.5 (The higher the age, the lesser the performance)***
